## 关于 QEP

**QEP (Quantum Error Processing)** 是由百度研究院[量子计算研究所](https://quantum.baidu.com)基于[量易伏](https://quantum-hub.baidu.com/)平台研发的量子噪声处理工具集，旨在从软件层面处理近期量子设备中广泛存在的量子噪声。现阶段该工具集提供了三大量子噪声处理功能：随机基准测试，量子噪声刻画，以及量子噪声缓释。

+ **随机基准测试**用于高效评估量子计算机的性能，衡量量子计算机可扩展性，包含随机基准测试、交替式基准测试、交叉熵基准测试、以及酉度随机基准测试等基础工具。

+ **量子噪声刻画**用于完整刻画量子计算机的量子噪声，精确标定量子态和量子门信息，包含量子态层析、量子过程层析、量子探测器层析、量子门集层析、以及量子谱层析等基础工具。

+ **量子噪声缓释**用于显著提升量子计算机的计算结果精度，消除量子门和量子测量的噪声，扩展量子设备的计算能力。包含零噪声外插量子门噪声缓释方法，矩阵逆方法、最小二乘法、迭代贝叶斯、Neumann 序列等多种量子噪声缓释方法。

## 常见问题

**Q: 使用 QEP 需要安装哪些依赖项?**

**A:** QEP 是基于量易伏 [QCompute](https://quantum-hub.baidu.com/opensource) 开源 SDK 开发，所以在使用 QEP 之前需要预先安装 QCompute。当用户安装 QEP 时会自动下载安装这个关键依赖包，其他所需的依赖项如下:

```bash
matplotlib==3.5.2
networkx==2.8.3
numpy==1.21.3
pandas==1.4.3
qiskit==0.37.0
qiskit_ibmq_provider==0.19.2
scipy==1.7.1
seaborn==0.11.2
tqdm==4.64.0
qutip==4.7.0
```

为了更好的用户体验，我们建议在使用 QEP 前将依赖项更新到最新版本。
